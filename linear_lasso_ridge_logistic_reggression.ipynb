```python
import pandas as pd
## laod data set and also we create the X and Y for our problem to  solve it by linear reggresion .
# Load the California housing dataset from an alternative source (CSV file)
data = pd.read_csv("./winequality-red.csv")
df = pd.DataFrame(data)
df
# Display the first few rows of the dataset
```

```python
X = df.iloc[:, :-1]
X
```

```python
Y = df.iloc[:,-1]
Y
## number of fetures in this problem is  => 11 
## the output is a number between 3 and 8
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pandas as pd

#########################  Linear reggression    ###################

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score

y_predict = model.predict(X_train)
y_train
mse = mean_squared_error(y_train,y_predict)
mse
r2_score_result = r2_score(y_train,y_predict)
print(f"model coefetiants => {model.coef_}")
print(f"modle r2_score result => {r2_score_result}")
print(f"MSE=> {mse}")

```

```python
#########################  Ridge reggression    ###################
from sklearn.linear_model import Ridge


# Create a linear regression model
model = Ridge(alpha=1)
# Fit the model to the training data
model.fit(X_train, y_train) 

y_predict = model.predict(X_train)
y_train
mse = mean_squared_error(y_train,y_predict)
mse
r2_score_result = r2_score(y_train,y_predict)
print(f"model coefetiants => {model.coef_}")
print(f"modle r2_score result => {r2_score_result}")
print(f"MSE=> {mse}")
```

```python
#########################  Ridge reggression    ###################
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

# Create a linear regression model
model = Lasso(alpha=1)
# Fit the model to the training data
model.fit(X_train, y_train) 

y_predict = model.predict(X_train)
y_train
mse = mean_squared_error(y_train,y_predict)
mse
r2_score_result = r2_score(y_train,y_predict)
print(f"model coefetiants => {model.coef_}")
print(f"modle r2_score result => {r2_score_result}")
print(f"MSE=> {mse}")
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV

######################### hyper tunne the model with the alpha value ########################

# Define a range of alpha values to search

alpha_values = [0.01, 0.1, 1, 10, 100,200]

# Create a Ridge regression model
model = Ridge()

# Define hyperparameters to tune
param_grid = {'alpha': alpha_values}

# Initialize GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best model and its hyperparameters
best_model = grid_search.best_estimator_
best_alpha = best_model.alpha

# Print the best hyperparameters
print(f"Best alpha: {best_alpha}")

# Fit the best model to the training data
best_model.fit(X_train, y_train)

# Predict on the training data
y_predict = best_model.predict(X_train)

# Calculate mean squared error
mse = mean_squared_error(y_train, y_predict)

# Calculate R-squared score
r2_score_result = r2_score(y_train, y_predict)

best_params = grid_search.best_params_
# Print results

print(f"Model coefficients: {best_model.coef_}")
print(f"Model R-squared score: {r2_score_result}")
print(f"MSE: {mse}")
print("Best Parameters:", best_params)
```

```python
######################### logistic regression ########################

from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame for easier handling
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Display the first few rows of the dataset
print("Iris Dataset:")
df
```

```python
iris_X = df.iloc[:,:-1]
iris_Y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import LogisticRegression

model_logistic_regression = LogisticRegression() 
model_logistic_regression.fit(iris_X,iris_Y)
```

```python
iris_Y_predected = model_logistic_regression.predict(iris_X)

# Calculate mean squared error
mse = mean_squared_error(iris_Y, iris_Y_predected)

# Calculate R-squared score
r2_score_result = r2_score(iris_Y, iris_Y_predected)

# Print results

print(f"Model R-squared score: {r2_score_result}")
print(f"MSE: {mse}")
```
