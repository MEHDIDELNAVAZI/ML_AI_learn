```python
import matplotlib as plt 
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
```

```python
load_data = pd.read_csv("./play_tennis.csv")

label_encoder = LabelEncoder()
load_data['Outlook']  = label_encoder.fit_transform(load_data['Outlook'])
load_data['Temperature']  = label_encoder.fit_transform(load_data['Temperature'])
load_data['Humidity']  = label_encoder.fit_transform(load_data['Humidity'])
load_data['Wind']  = label_encoder.fit_transform(load_data['Wind'])
load_data['Play Tennis']  = label_encoder.fit_transform(load_data['Wind'])
load_data = pd.DataFrame(load_data)
```

```python

X = load_data.iloc[:,:-1]
Y = load_data.iloc[:,-1]
```

```python
Y
```

```python
model = DecisionTreeClassifier()
model.fit(X,Y)
```

```python
y_predicted = model.predict(X)
mse = mean_squared_error(Y,y_predicted)
y_predicted
```

```python
from sklearn import tree

# Visualize the decision tree
tree.plot_tree(model, filled=True)
```
