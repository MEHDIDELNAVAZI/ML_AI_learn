```python
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import LabelEncoder

X = df.iloc[:,:-1]
Y = df.iloc[:,-1]


label_encoder = LabelEncoder()
X['Outlook']  = label_encoder.fit_transform(X['Outlook'])
X['Temperature']  = label_encoder.fit_transform(X['Temperature'])
X['Humidity']  = label_encoder.fit_transform(X['Humidity'])
X['Wind']  = label_encoder.fit_transform(X['Wind'])

X
```

```python
Y  = label_encoder.fit_transform(Y)
Y
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import mean_squared_error
# Build a Gaussian Classifier
model = GaussianNB()
# Model training
model.fit(X, Y)

Y_predeted=model.predict(X)
mse = mean_squared_error(Y,Y_predeted)
mse
```
